```r
# install packages
install.packages("MARSS")
install.packages("readxl")
install.packages("xtable")
library("MARSS")
library("readxl")
library("xtable")
```

```r
#import data
DFA_data <- read_excel("/content/Indicators.xlsx")
DFA_data
```

```r
#Check rows and columns for confirmation of correct data imported
Rows <- nrow(DFA_data)
Columns <- ncol(DFA_data)
Rows
Columns
```

```r
#Transpose data
DFA_dataT <- t(DFA_data)
DFA_dataT
```

```r
#standardize data 
dataZ <-  zscore(DFA_dataT)
dataZ
```

```r
#Fit the data to model

big.maxit.cntl.list <- list(minit = 200, maxit = 5000, allow.degen = FALSE)
model.list <- list(m = 1, R = "unconstrained")
the.fit <- MARSS(DFA_dataT, model = model.list, form = "dfa", 
                   control = big.maxit.cntl.list)
```

```r
#The model
trend <- the.fit$states
trend
write.csv(trend, file="activity.csv")
write.csv(the.fit$ytT,file="matrix.csv")
write.csv(the.fit$coef,file="coef.csv")
write.csv(Z,file="Z.csv")
Z <- coef(the.fit, type = "matrix")$Z
Z
```

```r
#plot trend to check if the model is correct
trendT <- t(trend)
x <- c(1:Rows)
plot(x=x,y=trendT,type="l")
```
